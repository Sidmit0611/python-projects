```python
# pandas is used to read the excel sheet and to manupulate with the data
import pandas as pd

# python uses SMTP protocol to send the mails that is why sptplib is import and from that all the functions are imported
from smtplib import *

# mimemultipart allows us to attach things with our email
# weather it is a plain text or HTML text, mimemultipart text is used
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

#here we are using gmail to send the mail that is why we have used smtp.gmail.com in our host
#for some person if using other service like yahoo and other might have different host
host = "smtp.gmail.com"

# port number 587 is assigned to gmail
port = 587

# username contains the e-mail id of the sender
username = "siddhantmittal1001@gmail.com"

# password contains the password of the sender email id
password = '**********'

# variable from e-mail contains the sender's email adress
from_email = username

# data contains the information of the persons whom you want to send the mail
# the data contains name and the email of the person
# to see the data refer the upper cell
data = pd.read_excel('Book1.xlsx')

# this line reads only the email id's of the persons(list comprehension is used)
ids = data.iloc[:, 1].values
list_to_send = [] # we will append the email id's of the persons into this empty list

for i in ids:
    list_to_send.append(str(i))

to_list = list_to_send # to list should be of list type which should contain the email id as string format
                       # that is why upper for loop is used

# all the main code is added into try block so that we can catch the exception if there is any exception
try:
    email_conn = SMTP(host, port) # is used to set up the email connection 
    email_conn.ehlo()             # will return the response of connection if connected(necessary step)
    email_conn.starttls()         # helps in encryption (simply we can understand, it is used to start the connection) it will return ready to start
    email_conn.ehlo()             
    email_conn.login(username, password)  # used to login into gmail using username and password( it will return you the message of 'Accepted')
    the_msg = MIMEMultipart("alternative") #alternative is a ordinary way to add the html msg
    the_msg['Subject'] = 'Program send through code in python'  # contains the subject you want to enclose with your mail
    the_msg['From'] = from_email          # sender's email adress
    
    
    # plain text can contain the empty string or string of you choice
    # our html text is send with the plain text
    plain_txt = " "
    
    # the html text we are using which contains the link of my linked in profile attached with my name
    html_txt = """\
    <html>
        <head> </head>
        <body>
        <p><h1>Hey!</h1><br>
             <i>testing the mail sercive via <b>Python</b> made by</i> <i><u><a href = 'https://www.linkedin.com/in/siddhant-mittal-7636a8148/'> <b> Siddhant Mittal </b></a></i></u>
        </p>
        </body>
     </html>
        """
    part_1 = MIMEText(plain_txt, 'plain')
    part_2 = MIMEText(html_txt, "html")
    
    # plain text attached with the message
    the_msg.attach(part_1)
    
    # html text attached with the message
    the_msg.attach(part_2)
    
    # sendmail is used to send the mail
    # requires 3 attribute('sender_email', ['email_of_person_you_want_to_send'], 'message')
    email_conn.sendmail(from_email, to_list, the_msg.as_string())
    
    # is used to quit the connection
    # we have to quit the connection after sending the mail
    email_conn.quit()
except smtplib.SMTPException:
    print("error accured")
```
